````python

def download_youtube_audio(url, output_path, preferred_quality="192"):
  ydl_opts = {
      'format': 'bestaudio/best',  # Select best audio quality
      'postprocessors': [{
          'key': 'FFmpegExtractAudio',
          'preferredcodec': 'mp3',
          'preferredquality': preferred_quality,
      }],
      'outtmpl': output_path,  # Specify the output path and file name
  }

  try:
      with youtube_dl.YoutubeDL(ydl_opts) as ydl:
          info_dict = ydl.extract_info(url, download=False)
          video_title = info_dict.get('title', None)
          print(f"Downloading audio for: {video_title}")

          ydl.download([url])
          print(f"Audio file saved as: {output_path}")

      return output_path

  except youtube_dl.utils.DownloadError as e:
      print(f"Error downloading audio: {e}")
      return None  # Indicate failure
````

````python
from typing import Generator
from faster_whisper import WhisperModel
import pandas as pd
model = WhisperModel("distil-large-v2")

path = "test"
download_youtube_audio("https://www.youtube.com/watch?v=Sby1uJ_NFIY", path)

path = path+".mp3"

def transcribe(path ,model):
    model = WhisperModel(model)

    print(f"reading {path}")
    segments, info = model.transcribe(path)
    return segments



def process_segments(segments: Generator):
    result = {}
    print("processing...")
    for i, segment in enumerate(segments):
        chunk_id = f"chunk_{i}"
        
        result[chunk_id] = {
            'chunk_id': segment.id,
            'chunk_length': segment.end - segment.start,
            'text': segment.text,
            'start_time': segment.start,
            'end_time': segment.end
        }
    df = pd.DataFrame.from_dict(result, orient='index')

    return df

df = process_segments(transcribe(download_youtube_audio("https://www.youtube.com/watch?v=Sby1uJ_NFIY", path), "distil-large-v3"))
````

# Already chunked setences (through vad) are rechunked (combined) with the "_chunk" function of the semantic_chunkers library, whcich groups chunks based on semantic similarity 

# here deciding the parameters    (min_split_tokens=30, max_split_tokens=40, window_size=2)  is very crucial as they can make a lot of difference in the output , by experiment i found out that between 30 and 40 speakers speak for 15s most of the time so approx 2.33 tokens/second

````python
import pandas as pd
df = pd.read_csv("data.csv")
lis = list(df.text)
from semantic_chunkers import StatisticalChunker
from semantic_router.encoders import HuggingFaceEncoder
encoder = HuggingFaceEncoder(name = "sentence-transformers/all-MiniLM-L6-v2")
chunker = StatisticalChunker(    encoder=encoder,
    dynamic_threshold=True,
    min_split_tokens=30,
    max_split_tokens=40,
    window_size=2,
    enable_statistics=False,  # to print chunking stats
                            )
chunks = chunker._chunk(lis)
````

````python
len(chunks)
````

## chunk[i].splits gives a list containing all the docs (whisper chunks) in the semantic chunk

````python
row_index = 0
for i in range(len(chunks)):
    for j in range(len(chunks[i].splits)):
        df.at[row_index, 'chunk_id2'] = f'chunk_{i}'
        row_index += 1
````

## Now the whisper chunks are grouped based on semantic chunks (all the whisper-chunks having same semantic chunk become one value based on the logic

- start-time will be minimum of all strt times
- end-time will be max (hope this makes sense)
- text column joins all the strings 

````python
grouped = df.groupby('chunk_id2').agg({
    'start_time': 'min',
    'end_time': 'max',
    'text': lambda x: ' '.join(x),
    'chunk_id': list
}).reset_index()

# Rename the columns if needed
grouped = grouped.rename(columns={'chunk_id': 'chunk_ids'})
grouped['chunk_length'] = grouped['end_time'] - grouped['start_time']
grouped['chunk_id'] = grouped['chunk_id2']
grouped = grouped.drop(columns=['chunk_id2', 'chunk_ids' ])
grouped.to_csv('final.csv')
````

````python
import gradio as gr 
import pandas as pd 
import yt_dlp

import os

df = pd.read_csv("./final.csv")
transcripts = df.to_dict(orient='records')

# Function to download video using yt-dlp and generate transcript HTML
def download_video(youtube_url):
    # Define download options
    ydl_opts = {
        'format': 'mp4',
        'outtmpl': 'downloaded_video.%(ext)s',
        'quiet': True
    }

    # Extract video ID to check if already downloaded
    with yt_dlp.YoutubeDL({'quiet': True}) as ydl:
        info_dict = ydl.extract_info(youtube_url, download=False)
        video_ext = info_dict.get('ext')
        video_path = f'downloaded_video.mp4'

    # Check if video already downloaded
    if not os.path.exists(video_path):
        # Download the video
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([youtube_url])

    # Generate HTML for the transcript
    transcript_html = ""
    for t in transcripts:
        transcript_html += f'<div class="transcript-block"><a href="#" onclick="var video = document.getElementById(\'video-player\').querySelector(\'video\'); video.currentTime={t["start_time"]}; return false;">' \
                           f'[{t["start_time"]:.2f} - {t["end_time"]:.2f}]<br>{t["text"]}</a></div>'
    
    return video_path, transcript_html

# Function to search the transcript
def search_transcript(keyword):
    search_results = ""
    for t in transcripts:
        if keyword.lower() in t['text'].lower():
            search_results += f'<div class="transcript-block"><a href="#" onclick="var video = document.getElementById(\'video-player\').querySelector(\'video\'); video.currentTime={t["start_time"]}; return false;">' \
                              f'[{t["start_time"]:.2f} - {t["end_time"]:.2f}]<br>{t["text"]}</a></div>'
    return search_results

# CSS for styling
css = """
.fixed-video { width: 480px !important; height: 270px !important; }
.fixed-transcript { width: 480px !important; height: 270px !important; overflow-y: auto; }
.transcript-block { margin: 10px 0; padding: 10px; border: 1px solid #ddd; border-radius: 5px; background-color: #f9f9f9; }
.transcript-block a { text-decoration: none; color: #007bff; }
.transcript-block a:hover { text-decoration: underline; }
"""

# Gradio interface
with gr.Blocks(css=css) as demo:
    gr.Markdown("# YouTube Video Player with Clickable Transcript")

    with gr.Row():
        youtube_url = gr.Textbox(label="YouTube URL", placeholder="Enter YouTube video link here")
        download_button = gr.Button("Download and Display Transcript")
    
    with gr.Row():
        video = gr.Video(label="Video Player", elem_id="video-player", elem_classes="fixed-video")
        transcript_display = gr.HTML(label="Transcript", elem_classes="fixed-transcript")

    with gr.Row():
        search_box = gr.Textbox(label="Search Transcript", placeholder="Enter keyword to search")
        search_button = gr.Button("Search")
        search_results_display = gr.HTML(label="Search Results", elem_classes="fixed-transcript")

    # On button click, download the video and display the transcript
    def display_transcript(youtube_url):
        video_path, transcript_html = download_video(youtube_url)
        # Ensure the video path is correctly passed to the Gradio video component
        return video_path, transcript_html

    download_button.click(fn=display_transcript, inputs=youtube_url, outputs=[video, transcript_display])

    # On search button click, search the transcript and display results
    search_button.click(fn=search_transcript, inputs=search_box, outputs=search_results_display)

# Launch the interface
demo.launch()
````

````python
%%writefile app.py
from playwright.async_api import async_playwright, TimeoutError
import asyncio
import requests
import os
import json
from bs4 import BeautifulSoup
import time

base_url = "https://live.bible.is/bible/HINOHC/MAT/1"
req_url = []
req_head = []
max_refresh_attempts = 5

# Set the folder where the audio file will be saved
audio_folder = "bibleaudio"
json_folder = "biblejson"
os.makedirs(audio_folder, exist_ok=True)
os.makedirs(json_folder, exist_ok=True)
json_file = "text.json"


# Function to handle requests and capture audio file URL
async def playlist(request):
    try:
        if ".mp3" in request.url:
            req_url.append(request.url)
            req_head.append(request.headers)
    except Exception as e:
        print(f"Error capturing audio URL: {e}")


# Function to download audio file
async def audio_downloader(audio_filename):
    #audio_filename = f"{audio_folder}/1ESV.mp3"
    try:
        response = requests.get(req_url[0], headers=req_head[0])
        with open(audio_filename, "wb") as f:
            f.write(response.content)
            print(f'audiofile {audio_filename} saved')
    except Exception as e:
        print(f"Error downloading audio: {e}")


def find_script(content, timeout=10, poll_interval=1):
    print("in find_script func")
    start_time = time.time()
    script = None
    while script is None:
        script = BeautifulSoup(content, "html.parser").find(
            "script", {"id": "__NEXT_DATA__"}
        )
        if script is not None:
            break
        if time.time() - start_time > timeout:
            raise TimeoutError("Timeout waiting for script element.")
        time.sleep(poll_interval)  # Wait before polling again
    return script


# Function to parse JSON and extract text
async def json_parser(content, json_file=json_file):
    # Maximum number of refresh attempts
    max_refresh_attempts = 5
    # Maximum wait time for the script to appear
    max_wait_time = 10  # seconds

    for attempt in range(max_refresh_attempts):
        try:
            script = await asyncio.wait_for(
                asyncio.to_thread(find_script, content), timeout=max_wait_time
            )
            print("script", script)
            if script:
                # If the script is found, parse its content
                print("here")
                script_content = script.string
                data = json.loads(script_content)
                chapter_text_list = data["props"]["pageProps"]["chapterText"]
                paragraph = " ".join(
                    item["verse_text"]
                    for item in chapter_text_list
                    if "verse_text" in item
                )

                paragraph = paragraph.replace("\u201c", "").replace("\u201d", "")

                # Extract "book_id" and "chapter" attributes from any element in the list
                if chapter_text_list:
                    first_item = chapter_text_list[0]
                    book_id = first_item.get("book_id", "unknown_book")
                    chapter = first_item.get("chapter", "unknown_chapter")
                else:
                    book_id = "unknown_book"
                    chapter = "unknown_chapter"

                # Create the new filename
                new_json_file = f"{json_folder}/{book_id}_{chapter}.json"

                # Save the modified content to the new file
                with open(new_json_file, "w") as file:
                    json.dump({"paragraph": data}, file, indent=4)

                print(f"Script found and JSON content saved to {new_json_file}")
                return {"paragraph": paragraph}
        except Exception as e:
            print(f"Error getting script content: {e}")

    # If the script is not found after all attempts, return None or raise an exception
    return None


# Function to run Playwright and handle scraping
async def run(playwright):
    try:
        firefox = playwright.firefox
        browser = await firefox.launch(
            headless=True
        )  # Set headless to False for debugging
        context = await browser.new_context()
        page = await context.new_page()

        # Subscribe to "request" events.
        page.on("request", lambda request: playlist(request))

        # Navigate to the main page
        print("Navigating to the main page")
        await page.goto(base_url)
        await page.wait_for_load_state("domcontentloaded")
        print("Main page loaded")

        # Click on the dropdown to show the chapter links
        print("Clicking the dropdown button")
        await page.click("#chapter-dropdown-button.book-chapter")

        # Locate the "New Testament" heading
        new_testament_heading = await page.locator(
            "div.testament-title", has_text="New Testament"
        ).element_handle()
        if not new_testament_heading:
            print("New Testament heading not found")
            return

        # Get the position of the New Testament heading in the DOM
        new_testament_position = await new_testament_heading.evaluate(
            "node => Array.from(node.parentElement.children).indexOf(node)"
        )

        # Locate the parent element containing the "New Testament" heading
        parent_element = await new_testament_heading.evaluate_handle(
            "node => node.parentElement"
        )

        # Find all book-button elements within the parent element
        book_buttons = await parent_element.query_selector_all("div.book-button")

        # Filter book buttons to only those appearing after the New Testament heading
        new_testament_books = []
        for button in book_buttons:
            button_position = await button.evaluate(
                "node => Array.from(node.parentElement.children).indexOf(node)"
            )
            if button_position > new_testament_position:
                new_testament_books.append(button)
                button_html = await button.evaluate("node => node.innerHTML")
                # print("Button HTML:", button_html)

        book_ids = [await book.get_attribute("id") for book in new_testament_books]
        print("New Testament Books:", book_ids)

        # Collect all chapter URLs before navigating
        chapters_to_scrape = ["MAT"]
        chapters_to_scrape.append("https://live.bible.is/bible/EN1ESV/MAT/3")
        for book in new_testament_books:
            await book.scroll_into_view_if_needed()  # Ensure the element is in view
            chapter_links = await book.query_selector_all(
                "span.chapter-container.inactive-book-chapters a.chapter-box"
            )

            for chapter_link in chapter_links:
                chapter_href = await chapter_link.get_attribute("href")
                if chapter_href:
                    chapter_url = f"{base_url}{chapter_href}"
                    chapters_to_scrape.append(chapter_url)
                    print(chapter_url)

        # Maximum number of refresh attempts
        max_refresh_attempts = 5

        for chapter_url in chapters_to_scrape:
            try:
                # Navigate to each chapter URL and process the content
                await page.goto(chapter_url)
                await page.wait_for_load_state("domcontentloaded")
                # Getting page content
                content = await page.content()
                print("Page Loaded")

                await json_parser(content)
                parts = chapter_url.split('/')
                book_code = parts[-2]
                chapter_number = parts[-1]
                audio_filename  = f"bibleaudio/{book_code}_{chapter_number}.mp3"
                await audio_downloader(audio_filename)

            except Exception as e:
                print(f"Error processing chapter URL {chapter_url}: {e}")

            # Go back to the main page to select the next chapter
            # await page.goto(base_url)
            # await page.wait_for_load_state("domcontentloaded")
            # await page.click("#chapter-dropdown-button.book-chapter")
    except Exception as e:
        print(f"Error in Playwright run function: {e}")
    finally:
        await browser.close()


# Main execution
async def main():
    async with async_playwright() as playwright:
        await run(playwright)


# Run the script
asyncio.run(main())

# Run the script
asyncio.run(main())
````

````python
import pandas as pd
import glob
import os

# Define the directories
dir1 = '/kaggle/working/biblejson'
dir2 = '/kaggle/working/bibleaudio'

# Get all JSON and CSV files from both directories
files1 = glob.glob(os.path.join(dir1, '*'))
files2 = glob.glob(os.path.join(dir2, '*'))

# Extract file base names (without extensions) and full paths
files1_base = {os.path.splitext(os.path.basename(file))[0]: file for file in files1}
files2_base = {os.path.splitext(os.path.basename(file))[0]: file for file in files2}

# Find common base names
common_base_names = set(files1_base.keys()) & set(files2_base.keys())

# Initialize an empty list to hold data frames
data_frames = []

# Loop through the common base names and read the corresponding files into data frames
for base_name in sorted(common_base_names):
    df_list = []

    # Read the file from the first directory
    file1 = files1_base.get(base_name)
    if file1:
        if file1.endswith('.json'):
            try:
                df = pd.read_json(file1)
                df_list.append(df)
            except Exception as e:
                print(f"Error reading {file1}: {e}")
        elif file1.endswith('.csv'):
            try:
                df = pd.read_csv(file1)
                df_list.append(df)
            except Exception as e:
                print(f"Error reading {file1}: {e}")

    # Read the file from the second directory
    file2 = files2_base.get(base_name)
    if file2:
        if file2.endswith('.json'):
            try:
                df = pd.read_json(file2)
                df_list.append(df)
            except Exception as e:
                print(f"Error reading {file2}: {e}")
        elif file2.endswith('.csv'):
            try:
                df = pd.read_csv(file2)
                df_list.append(df)
            except Exception as e:
                print(f"Error reading {file2}: {e}")

    # Concatenate data frames from both directories for the current base name
    if df_list:
        combined_df = pd.concat(df_list, ignore_index=True)
        combined_df['audio_file_path'] = file1 if file1 else file2
        data_frames.append(combined_df)
    else:
        print(f"No valid data frames for base name {base_name}")

# Check if there are any data frames to concatenate
if data_frames:
    # Concatenate all combined data frames
    final_combined_df = pd.concat(data_frames, ignore_index=True)
    # Save the final combined data frame to a CSV file
    final_combined_df.to_csv('combined_output.csv', index=False)
    print("All files with the same base name have been combined and saved to 'combined_output.csv'.")
else:
    print("No data frames to concatenate. Please check your input files and directories.")
````

````python
from indicnlp.tokenize import indic_tokenize
from indicnlp.normalize.indic_normalize import IndicNormalizerFactory
import string

# Initialize normalizer
normalizer_factory = IndicNormalizerFactory()
normalizer = normalizer_factory.get_normalizer("hi")

# Tokenize and normalize text
def preprocess_text(text):
    tokens = indic_tokenize.trivial_tokenize(text)
    normalized_tokens = [normalizer.normalize(token) for token in tokens]
    return [token for token in normalized_tokens if token not in stopwords and token not in string.punctuation]

# Example text
text = "à¤­à¤¾à¤°à¤¤ à¤•à¥€ à¤°à¤¾à¤œà¤§à¤¾à¤¨à¥€ à¤¨à¤ˆ à¤¦à¤¿à¤²à¥à¤²à¥€ à¤¹à¥ˆà¥¤ à¤¯à¤¹à¤¾à¤ à¤ªà¤° à¤•à¤ˆ à¤ªà¥à¤°à¤®à¥à¤– à¤¦à¤°à¥à¤¶à¤¨à¥€à¤¯ à¤¸à¥à¤¥à¤² à¤¹à¥ˆà¤‚à¥¤"

# Preprocess the text
processed_text = preprocess_text(text)
print(processed_text)
````

````python
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt

# Create a dictionary from the preprocessed text
dictionary = Dictionary([processed_text])

# Create a corpus
corpus = [dictionary.doc2bow(processed_text)]

# Example: Print the corpus
print(corpus)
````

````python
# Train the LDA model
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=10)

# Print the topics
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
````

````python
# Compute coherence score
coherence_model_lda = CoherenceModel(model=lda_model, texts=[processed_text], dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print(f'Coherence Score: {coherence_lda}')
````

````python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt

# Vectorize the text
vectorizer = CountVectorizer(stop_words=stopwords)
X = vectorizer.fit_transform(data['verse_text'])

# Fit LDA model
n_topics = 5
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(X)

# Get the topic-word distribution matrix
topic_words = lda.components_
vocab = vectorizer.get_feature_names_out()

# Plot the top words in each topic
n_top_words = 10
fig, axes = plt.subplots(1, n_topics, figsize=(20, 5), sharex=True)
axes = axes.flatten()

for topic_idx, topic in enumerate(topic_words):
    top_features_ind = topic.argsort()[: -n_top_words - 1 : -1]
    top_features = [vocab[i] for i in top_features_ind]
    weights = topic[top_features_ind]

    ax = axes[topic_idx]
    ax.barh(top_features, weights, height=0.7)
    ax.set_title(f'Topic {topic_idx + 1}', fontproperties=hindi_font, fontdict={'fontsize': 15})
    ax.invert_yaxis()
    ax.tick_params(axis='both', which='major', labelsize=12)
    for i in 'top right left'.split():
        ax.spines[i].set_visible(False)

plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
plt.show()
````

````python
from indicnlp.tokenize import indic_tokenize
from collections import Counter
import itertools

# Tokenization
data['tokens'] = data['verse_text'].apply(lambda x: list(indic_tokenize.trivial_tokenize(x)))

# N-gram Analysis
def get_ngrams(tokens, n):
    ngrams = zip(*[tokens[i:] for i in range(n)])
    return [' '.join(ngram) for ngram in ngrams]

# Getting bigrams and trigrams
data['bigrams'] = data['tokens'].apply(lambda x: get_ngrams(x, 2))
data['trigrams'] = data['tokens'].apply(lambda x: get_ngrams(x, 3))

# Count bigrams and trigrams
bigrams = list(itertools.chain(*data['bigrams'].tolist()))
trigrams = list(itertools.chain(*data['trigrams'].tolist()))

bigram_counts = Counter(bigrams)
trigram_counts = Counter(trigrams)

# Displaying the most common bigrams and trigrams
print("Most common bigrams:")
print(bigram_counts.most_common(10))

print("\nMost common trigrams:")
print(trigram_counts.most_common(10))
````

````python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

# Load the tokenizer and model for NER
tokenizer = AutoTokenizer.from_pretrained("Davlan/bert-base-multilingual-cased-ner-hrl")
model = AutoModelForTokenClassification.from_pretrained("Davlan/bert-base-multilingual-cased-ner-hrl")
model.to("cuda")
# Initialize the NER pipeline
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer, device="cuda")

# Function to perform NER
def get_ner(text):
    ner_results = ner_pipeline(text)
    return [(result['word'], result['entity']) for result in ner_results]

# Apply NER to each verse
data['ner'] = data['verse_text'].apply(get_ner)

# Display some examples
print(data[['verse_text', 'ner']].head())
````

````python
entity_counts = {}

for entities in data['ner']:
    for entity in entities:
        entity_text = entity[0]
        if entity_text not in entity_counts:
            entity_counts[entity_text] = 1
        else:
            entity_counts[entity_text] += 1

# Convert entity counts to a DataFrame for plotting
entity_df = pd.DataFrame(list(entity_counts.items()), columns=['Entity', 'Count'])
entity_df = entity_df.sort_values(by='Count', ascending=False).head(20)  # Top 20 entities
````
